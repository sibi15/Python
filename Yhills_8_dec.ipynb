##### -> types of data: categorical and numerical

##### Numerical data -> mean| median| mode
##### Categorical data -> mode

##### Statistics: 
##### 1. Descriptive (Dataset -> description/explanation)

##### 2. Inferential (Population -> take small portion of data and work with that)

##### => Measures of Central Tendency: (3 M's -> Mean, median, mode)
##### Mean -> not a very good measure because it is impacted when there are extreme outliers, eg: 1,2,3,4,100
##### Median -> first sort the array then choose middle value (better than mean as a data scientist because it is not impacted by outliers)
##### Mode -> Maximum frequency element; problem when multiple elements with same modes

##### => Measures of dispersion: (spread of data)
##### Range -> Max vale - Min value
##### Variance -> (sigma^2) == sigma ( i=1 -> n) [( x - x bar )^2] / n
##### Standard Deviation (sigma) == (root (variance))

##### => Covariance and correlation: (relation between attributes)
##### -> Covariance == sigma (i=1 -> n) [( x - x bar )( y - y bar)] / n
##### if both attributes change in same direction -> positive covariance else negative
##### -> Correlation == covariance (x,y) / [std_deviation(x) - std_deviation(y)]
##### Correlation belongs to {-1,1}
##### ==> Covariance tells direction of relationship and correlation tells the strength of the relationship
